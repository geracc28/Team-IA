```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris, load_wine, load_breast_cancer
import numpy as np
import pandas as pd

def hold_out_stratified(dataset, test_size=0.3, random_state=None):
    X, y = dataset.data, dataset.target
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, stratify=y, random_state=random_state)
    return X_train, X_test, y_train, y_test

# Cargar los conjuntos de datos
iris = load_iris()
wine = load_wine()
breast = load_breast_cancer()

# Aplicar Hold Out 70/30 estratificado en los conjuntos de datos
iris_X_train, iris_X_test, iris_y_train, iris_y_test = hold_out_stratified(iris, test_size=0.3, random_state=42)
wine_X_train, wine_X_test, wine_y_train, wine_y_test = hold_out_stratified(wine, test_size=0.3, random_state=42)
breast_X_train, breast_X_test, breast_y_train, breast_y_test = hold_out_stratified(breast, test_size=0.3, random_state=42)

# Crear DataFrames para visualizar los resultados
iris_train_df = pd.DataFrame(data=iris_X_train, columns=iris.feature_names)
iris_test_df = pd.DataFrame(data=iris_X_test, columns=iris.feature_names)
wine_train_df = pd.DataFrame(data=wine_X_train, columns=wine.feature_names)
wine_test_df = pd.DataFrame(data=wine_X_test, columns=wine.feature_names)
breast_train_df = pd.DataFrame(data=breast_X_train, columns=breast.feature_names)
breast_test_df = pd.DataFrame(data=breast_X_test, columns=breast.feature_names)

# Mostrar los DataFrames en Jupyter
from IPython.display import display

print("---Iris---")
print("Entrenamiento")
display(iris_train_df)
print("Prueba")
display(iris_test_df)

print("---Wine---")
print("Entrenamiento")
display(wine_train_df)
print("Prueba")
display(wine_test_df)

print("---breast---")
print("Entrenamiento")
display(breast_train_df)
print("Prueba")
display(breast_test_df)
```
