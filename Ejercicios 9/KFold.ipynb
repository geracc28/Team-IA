```python
from sklearn.model_selection import StratifiedKFold
from sklearn.datasets import load_iris, load_wine, load_breast_cancer
import numpy as np
import pandas as pd

def stratified_kfold_cross_validation(dataset, n_splits=10, random_state=None):
    X, y = dataset.data, dataset.target
    skf = StratifiedKFold(n_splits=n_splits, random_state=random_state, shuffle=True)
    splits = []
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        splits.append((X_train, X_test, y_train, y_test))
    return splits

# Cargar los conjuntos de datos
iris = load_iris()
wine = load_wine()
breast = load_breast_cancer()

# Aplicar validación cruzada estratificada de 10-Fold en los conjuntos de datos
iris_splits = stratified_kfold_cross_validation(iris, n_splits=10, random_state=42)
wine_splits = stratified_kfold_cross_validation(wine, n_splits=10, random_state=42)
breast_splits = stratified_kfold_cross_validation(breast, n_splits=10, random_state=42)

# Mostrar un ejemplo de los conjuntos de datos generados para cada conjunto
datasets = [(iris_splits, iris), (wine_splits, wine), (breast_splits, breast)]
for dataset_splits, dataset_info in datasets:
    print("---", dataset_info['DESCR'].split('\n')[0], "---")
    for i in range(3):  # Solo mostraremos los primeros 3 de cada conjunto de datos
        print("Entrenamiento")
        display(pd.DataFrame(data=dataset_splits[i][0], columns=dataset_info.feature_names))
        print("Prueba")
        display(pd.DataFrame(data=dataset_splits[i][1], columns=dataset_info.feature_names))
```
